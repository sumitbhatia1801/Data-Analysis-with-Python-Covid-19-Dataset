### Remove all the records where confimed cases is less than 10.

```python
data.head()
```

```python
data = data[~(data["Confirmed"]<10)]
```

```python
data.shape
```

### In which region maximum number of confirmed cases were recorded?

```python
data.groupby(by = "Region")["Confirmed"].sum().sort_values(ascending = False)

# Answer - US
```

### In which region minimum number of deaths were recorded ?

```python
data.head()
```

```python
data.groupby(by = "Region")["Deaths"].sum().sort_values().head(30)
```
